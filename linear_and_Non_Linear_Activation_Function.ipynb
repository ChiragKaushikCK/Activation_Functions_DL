```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Linear Function (y = mx + b)
def linear_function(x, m=1, b=0):
    return m * x + b

# 2. ReLU Function (y = max(0, x))
def relu_function(x):
    return np.maximum(0, x)

# Generate values for x
x_values = np.linspace(-10, 20)

# Compute y values for both functions
linear_y_values = linear_function(x_values)
relu_y_values = relu_function(x_values)

# Plotting the functions
plt.figure(figsize=(10,3))

# Plot Linear Function
plt.subplot(1, 2, 1)
plt.plot(x_values, linear_y_values, label='Linear Function (y = mx + b)', color='black')
plt.title('Linear Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Plot ReLU Function
plt.subplot(1, 2, 2)
plt.plot(x_values, relu_y_values, label='ReLU Function (y = max(0, x))', color='red')
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

```python
#Sigmoid activation function
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate values for x
x_values = np.linspace(-10, 10)

#compute the value for function
sigmoid_y_values = sigmoid(x_values)

# Plotting the functions
plt.figure(figsize=(10,3))

# Plot Linear Function
plt.subplot(1, 2, 1)
plt.plot(x_values, sigmoid_y_values, label='Sigmoid Function (y = 1 / (1 + e^-x))', color='black')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))  # Change here

#generate value for function
x_values = np.linspace(-10, 10)
tanh_y_values = tanh(x_values)

# Plotting the functions
plt.figure(figsize=(10,3))

#plot linear function
plt.subplot(1, 2, 1)
plt.plot(x_values, tanh_y_values, label='Tanh Function (y = (e^x - e^-x) / (e^x + e^-x))', color='black')
plt.title('Tanh Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
```
